### So in this notebook we have train a Word2Vec from sratch just like google for sentiment analysis 

```python
import pandas as pd


```

```python
df=pd.read_csv("C:/Users/Amaan Shaikh/Downloads/all_kindle_review .csv")
```

```python
df
```

```python
df=df[["rating","reviewText"]]
```

```python
df
```

```python
df["rating"]=df["rating"].apply(lambda x:0 if x<3 else 1 )
```

```python
df
```

```python
corpus=[]
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
lem=WordNetLemmatizer()
```

```python
for i in range(len(df)):
    review=df["reviewText"][i]
    review=review.lower()
    review=review.split()
    review=[lem.lemmatize(word) for word in review]
    review=" ".join(review)
    corpus.append(review)
    
```

```python
corpus
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
model=Word2Vec(corpus,vector_size=100)
```

```python
model.corpus_count
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
X=[]
```

```python
for i in range(0,len(corpus)):
    X.append(avg_word2vec(corpus[i]))
    
```

```python
len(X)
```

```python
y=df["rating"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
ypred=rf.predict(x_test)
```

```python
print(classification_report(y_test,ypred))
```
